```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("weight-height.csv")
```

```python
df.head()
```

```python
sns.distplot(df["Height"])
```

```python
sns.boxplot(df["Height"])
```

```python
min_limit = df["Height"].quantile(0.01)
max_limit = df["Height"].quantile(0.99)
```

```python
min_limit
```

```python
max_limit
```

```python
df[(df["Height"]>max_limit) | (df["Height"]<min_limit)]
```

# Trimming

```python
new_data = df[(df["Height"]<max_limit) & (df["Height"]>min_limit)]
```

```python
sns.boxplot(new_data["Height"])
```

# Capping

```python
df["Height"] = np.where(df["Height"]>max_limit,max_limit, np.where(df["Height"]<min_limit,min_limit, df["Height"]))
```

```python
sns.boxplot(df["Height"])
```
