## Filling The Missing Values
(i) Numerical Data - use mean

```python
df1 =df.copy()
```

```python
mean_value = df1['reviews_per_month'].mean()
df1["reviews_per_month"].fillna(mean_value,inplace = True)     
# Mean Of The reviews_per_month column fill in the place of missing Values
```

```python
df1.head()
```

## Filling Categorical Data

```python
df1.nunique()
```

```python
df1["last_review"].fillna(df1["last_review"].value_counts().index[0],inplace = True)
```
